## The Activation Function

***activation function***은 단순히 각 계층의 출력에 적용하는 함수다. 가장 일반적인 함수는 Rectifier function이다. max(0,x)

***rectified linear unit*** , ***ReLU***  <br>
선형 유닛에 ReLU activation을 적용하면 출력이 max(0, w*x + b)가 되며 다음과 같음 다이어그램처럼 나타낼 수 있다. 


## Stacking Dense Layers

output layer 전에 오는 layer를 ***hidden***이라 부른다.

## Building Sequential Models

```python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    # the hidden ReLU layers
    layers.Dense(units=4, activation='relu', input_shape=[2]),
    layers.Dense(units=3, activation='relu'),
    # the linear output layer 
    layers.Dense(units=1),
])
```
