## Get out environment set up

```python
import pandas as pd
import numpy as np

from scipy import stats

from mlxtend.preprocessing import minmax_scaling

import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
```

## Scaling vs Normalization : 차이점?
<b>scaling</b>은 데이터의 범위를 바꾸는 것이다. <br>
<b>normalization</b>데이터의 분포(distribution) 형태를 바꾸는 것이다.

## Scaling
데이터가 0-100 또는 0-1과 같은 특정 척도 내에 적합하도록 변환한다는 것이다.<br>
변수의 척도를 scaling하면 서로 다른 변수를 동일한 기준으로 비교할 수 있다.


```python
original_data = np.random.exponential(size=1000)

scaled_data = minmax_scaling(original_data, columns=[0])

fig, ax = plt.subplots(1,2)
sns.distplot(original_data, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(scaled_data, ax=ax[1])
ax[1].set_title("Scaled data")
```

## Normalization
scaling은 데이터의 범위를 바꿨다. Normalization은 더 radical한 변형이다. Normaliztaion은 관측치를 정규 분포로 설명할 수 있도록 변경하는 것이다.

```python
normalized_data = stats.boxcox(original_data)
fig, ax = plt.subplots(1, 2)
sns.distplot(original_data, ax=ax[0])
ax[0].set_title("Original Data")
sns.distplot(normalized_data[0], ax=ax[1])
ax[1].set_title("Normalized data")
```
