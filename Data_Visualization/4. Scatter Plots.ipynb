````python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
````

## Load and examine the data
insurance charges 데이터셋을 사용할 것이다. 일부 고객이 다른 고객보다 더 많은 비용을 지불하는 이유를 이해할 수 있는지 확인한다.

````python
insurance_filepath = "../input/datavis/insurance.csv"

insurance_data = pd.read_csv(insurance_filepath)
````

````python
insurance_data.head()
````

## Scatter plots
```sns.scatterplot```명령어를 사용하여 ***scatter plot***을 생성할 것이다.

````python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
````

이 관계의 strength를 double-check하려면 ***regression line*** 또는 가장 적합한 선을 추가하는 것이 좋다. ```sns.regplot``` 명령어로 바꿔서 이를 표현할 수 있다.

````python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
````

## Color-coded scatter plots
scatter를 사용하여 세 변수 간의 관계도 표시할 수 있다.

````python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
````

비흡연자가 bmi가 증가함에 따라 비용을 많이 내는 것을 볼 수 있지만 흡연자의 경우 대부분 더 많은 금액을 내는 것을 확인할 수 있다. <br>
이 사실을 더 강조하기 위해 우리는 ```sns.lmplot``` 명령어를 사용할 수 있다. 이것은 두가지 regression lines을 추가한다.

````python
sns.lmplot(x="bmi",y="charges",hue="smoker",data=insurance_data)
````

더 나아가서 주 축(main axes) 중 하나가 범주형 변수(ex: smoker)가 포함되도록 scatter의 디자인을 조정할 수 있다. 이런 plot유형을 ***categorical scatter plot***이라 하고 ```sns.swarmplot```명령어로 작성한다.

````python
sns.swarmplot(x=insurance_data['smoker'],
             y=insurance_data['charges'])
````
