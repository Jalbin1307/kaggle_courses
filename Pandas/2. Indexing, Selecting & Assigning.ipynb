## Indexing in pandas
인덱싱 연산자와 속성 선택은 Python ecosystem의 나머지 부분과 동일하기 작동하여 초보자가 쉽게 사용할 수 있다.<br>
하지만 pandas는 ```loc```와 ```iloc```라는 독립적인 접근 연산자를 가지고 있다. 고급 작업을 위해 이러한 기능들을 사용해야한다.

### Index-based selection
```iloc```. 데이터의 numerical position에 따른 데이터 선택.

````python
reviews.iloc[0]
````

위와 같은 ```loc``` 과 ```iloc```의 출력은 행이 첫번째, 열이 두번째이다. 기존의 파이썬에서 출력과 반대된다.<br>
이것은 행을 검색하는 것이 약간 더 쉬워지고 열을 검색하는 것이 약간 더 어려워진다. iloc으로 열을 가져오려면 다음과 같이 수행한다.

````python
reviews.iloc[:,0]
````

````python
reviews.iloc[:3, 0]
````

````python
reviews.iloc[1:3, 0]
````

````python
reviews.iloc[[0, 1, 2], 0]
````

````python
reviews.iloc[-5:]
````

### Label-based selection
```loc```. 데이터 인덱스 값

````python
reviews.loc[0, 'country']
````

```iloc```은 데이터셋의 인덱스를 무시하므로 ```loc```보다 개념적으로 간단하다. <br>
iloc은 보통 데이터셋을 위치별로 인덱싱해야 하는 큰 매트릭스에 사용되고 loc은 인덱스들의 정보를 사용하여 작업을 수행한다.

````python
reviews.loc[:, ['taster_name', 'taster_twitter_handle', 'points']]
````

### Choosing between ```loc``` and ```iloc```

loc과 iloc을 사용할 때 유의해야할 점이 있다. 두 방법은 약간 다른 인덱싱 체계를 사용한다.
<br><br>
iloc에서는 첫번째 요소가 포함되고 마지막 요소가 제외되는 Python stdlib 인덱싱 체계를 사용한다. 0:10 이면 0부터 9까지 <br>
반면 loc에서는 모두 포함하는 인덱싱 체계이다. 0:10 이면 0부터 10까지<br><br>
loc은 stdlib 유형(예: 문자열)을 인덱싱 할 수 있다. 색인 값이 Apple, ..., Potatoes,... 인 프레임이 있고 "Apple과 Potatoes" 사이의 모든 알파벳 과일을 선택하려면 df.loc['Apple':'Potatoes']로 인덱싱하는 것이 훨씬 더 편리하다. <br><br>
Dataframe의 인덱스가 0,...,1000과 같은 단순한 숫자 list일 경우 혼란이 올 수 있다. df.loc[0:1000] 은 1001개의 항목을 반환하기 때문에 한번 더 작업을 해줘야한다 (df.loc[0:999])

## Conditional selection

````python
reviews.country == 'Italy'
````

````python
reviews.loc[reviews.country == 'Italy']
````

````python
reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 90)]
````

````python
reviews.loc[(reviews.country == 'Italy') | (reviews.points >= 90)]
````

````python
reviews.loc[reviews.country.isin(['Italy', 'France'])]
````

````python
reviews.loc[reviews.price.notnull()]
````
