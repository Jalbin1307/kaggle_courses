## What have I learned?




- ***Trends*** : trend는 패턴의 변화로 정의된다.
    - ```sns.lineplot``` - Line Charts는 기간의 trend를 보여주는데 좋고 여러 line들을 사용할 수 있다.
- ***Relationship*** : 데이터의 변수 간의 관계를 이해하는 데 사용할 수 있는 다양한 차트 유형
    - ```sns.barplot``` - ***Bar charts***는 여러 그룹에 해당하는 수량을 비교할 때 유용하다.
    - ```sns.heatmap``` - ***Heatmaps***는 숫자 테이블에서 color-coded 패턴을 찾을 수 있다.
    - ```sns.scatterplot``` - ***Scatter plots***는 두개의 연속적인 값 사이의 관계를 보여주고 color-coded인 경우 세번째 범주형 변수와의 관계도 표시할 수 있다.
    - ```sns.regplot``` - scatter에서 ***regression line***을 포함하면 두 변수 사이의 선형 관계를 더 쉽게 확인할 수 있다. 
    - ```sns.lmplot``` - scatter에 여러 color-coded 그룹이 포함된 경우 multiple regression lines을 그릴 때 유용하다. 
    - ```sns.swarmplot``` - ***Categorical scatter plots*** 연속적인 변수와 범주형 변수 사이의 관계를 보여준다.
- ***Distribution*** : 분포를 시각화하여 변수에서 예상할 수 있는 값과 함께 분포의 가능성을 보여준다.
    - ```sns.distplot``` - ***Histograms***은 단일(single) 숫자 변수의 분포를 보여준다.
    - ```sns.kdeplot``` - ***KDE plots*** (or ***2D KDE plots***)은 단일 숫자 변수(또는 두개의 숫자 변수)의 추정된 매끄러운 분포를 보여준다.
    - ```sns.jointplot``` - 이 명령은 각 개별 변수에 해당하는 KDE 그림이 있는 2D KDE 그림을 동시에 표시하는데 유용하다.

## Changing styles with seaborn

````python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
````

````python
spotify_filepath = "../input/datavis/spotify.csv"

spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)

plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
````

````python
sns.set_style("darkgrid")

plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
````

Seaborn은 (1)"darkgrid", (2)"whitegrid", (3)"dark", (4)"white", and (5)"ticks" 이렇게 총 5가지 테마가 있다.
